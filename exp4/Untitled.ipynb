```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
```

```python
g = 9.81
l = 1
```

```python
sol = odeint(lambda y, t: [y[1], -g/l * np.cos(y[0]) * y[0]], [np.pi/10, 0], np.linspace(0, 10, 50))
plt.plot(sol[:, 0])
```

```python
sol = odeint(lambda y, t: [y[1], -g/l * y[0]], [np.pi/10, 0], np.linspace(0, 10, 50))
plt.plot(sol[:, 0])
```
